# 1. Preparamos los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_excel('Datos Solaz.xlsx')
```

```python
df.head()
```

```python
#print the percentage of missing values in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Drop the rows with null values in the column 'Estado cognitivo'
df = df.dropna(subset=['Estado cognitivo'])
```

```python
#print the percentage of missing values in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Drop the rows with null values in the columns 'Barthel' and 'MMSE' simultaneously

df = df.dropna(subset=['Barthel', 'MMSE'], how = 'all') 
```

```python
#print the percentage of missing values in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.info()
```

```python
#drop unnecessary columns

df.drop(['Cedula', 'Fecha de nacimiento', 'Con quien vive', 'Quirúrgicos', 'Conducta',
         'Dx. Cie 10 1', 'Dx. Cie 10 2', 'Dx. Cie 10 3','Conceptualizacion diagnostica',
         'Categoría Solaz', 'Riesgo de caida','Unnamed: 134'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Replace numerical values with the corresponding categorical values

#Ocupación
ocupacion = {0:'', 1: 'Comercio', 2: 'Industria', 3: 'Servicios', 4: 'Construccion', 
             5: 'Transporte', 6: 'FFAA o policia', 7: 'Desempleado', 8: 'Ama de casa',
             9: 'Estudiante', 10: 'Pensionado(a)'}
df['Ocupación'] = df['Ocupación'].replace(ocupacion)

#Estado civil
estado_civil = {0:'', 1: 'Soltero', 2: 'Union libre', 3: 'Casado', 4: 'Separado',
                5: 'Divorciado', 6: 'Viudo'}
df['Estado civil'] = df['Estado civil'].replace(estado_civil)

#Bipedestacion
bipedestacion = {0: 'No', 1: 'Con ayuda', 2: 'Si'}
df['Bipedestacion'] = df['Bipedestacion'].replace(bipedestacion)

#Apoyo en talones
apoyo_en_talones = {0: 'No', 1: 'Con ayuda', 2: 'Si'}
df['Apoyo en talones'] = df['Apoyo en talones'].replace(apoyo_en_talones)

#Apoyo en puntas
apoyo_en_puntas = {0: 'No', 1: 'Con ayuda', 2: 'Si'}
df['Apoyo en puntas'] = df['Apoyo en puntas'].replace(apoyo_en_puntas)

#Estado cognitivo
estado_cognitivo = {0: 'Demencia', 1: 'Otra alteracion mental o del SNC', 2: 'Cognitivamente Sano'}
df['Estado cognitivo'] = df['Estado cognitivo'].replace(estado_cognitivo)


```

```python
#Replace categorical values with the corresponding numerical values
Fuerza = {'1': 0, '1+': 1, '2': 2, '2+': 3, '3': 4, '3+': 5, '4': 6, '4+': 7, '5': 8}
df['Fuerza'] = df['Fuerza'].replace(Fuerza)
```

```python
df.head()
```

```python
#Replace 0 with false and 1 with true in columns which contain 0, 1 or NaN

Non_binary_columns = ['Fecha de evaluacion', 'Edad', 'Escolaridad', 'Ocupación', 'Estado civil', 'Estado cognitivo',
                      'Lugar de nacimiento', 'Lugar de residencia', 'Número de hijos', 'Número de nietos',
                      'Numero de bisnietos', 'Numero de medicamentos' ,'Hace cuanto dejo de fumar',
                      'Barthel', 'MMSE','Saturacion de O2','Frecuencia respiratoria', 'Presión arterial sistólica',
                      'Presión arterial diastólica', 'Frecuencia cardiaca','Grupo sanguineo','Fuerza',
                      'Bipedestacion', 'Apoyo en talones', 'Apoyo en puntas']

#for column not in Non_binary_columns
for col in df.columns:
    if col not in Non_binary_columns:
        df[col] = df[col].replace({0: False, 1: True})
```

```python
 #graficamos los datos numericos
# Boxplot de las variables numéricas
for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        # Boxplot
        plt.figure()
        df[col].plot.box()
        plt.title(col)
        plt.show()
```

```python
 #violin plot de las variables numéricas
import seaborn as sns

for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        # Violinplot
        plt.figure()
        sns.violinplot(x = df[col])
        plt.title(col)
        plt.show()
```

```python
 #barh plot de las variables categoricas
for col in df.columns:
    if df[col].dtype == 'category' or df[col].dtype == 'object' or df[col].dtype == 'bool':
        #barh plot
        plt.figure()
        df[col].value_counts().plot(kind='barh')
        plt.title(col)
        plt.show()
```
