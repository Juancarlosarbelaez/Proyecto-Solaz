# Imputacion de nulos y atipicos


```python
df = pd.read_excel('Datos Solaz_cleaned.xlsx')
```

```python
df.head()
```

```python
#print the percentage of missing values in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#for each column print its type
for col in df.columns:
    print('{} - {}'.format(col, df[col].dtype))
```

```python
#Replace 0 with false and 1 with true in columns which contain 0, 1 or NaN

Non_binary_columns = ['Fecha de evaluacion', 'Edad', 'Escolaridad', 'Ocupación', 'Estado civil', 'Estado cognitivo',
                      'Lugar de nacimiento', 'Lugar de residencia', 'Número de hijos', 'Número de nietos',
                      'Numero de bisnietos', 'Numero de medicamentos' ,'Hace cuanto dejo de fumar',
                      'Barthel', 'MMSE','Saturacion de O2','Frecuencia respiratoria', 'Presión arterial sistólica',
                      'Presión arterial diastólica', 'Frecuencia cardiaca','Grupo sanguineo','Fuerza',
                      'Bipedestacion', 'Apoyo en talones', 'Apoyo en puntas']

#for column not in Non_binary_columns
for col in df.columns:
    if col not in Non_binary_columns:
        df[col] = df[col].replace({0: False, 1: True})
```

```python
#for each column print its type
for col in df.columns:
    print('{} - {}'.format(col, df[col].dtype))
```

```python
#transform object columns to category
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category')
```

```python
df.drop(['Fecha de evaluacion', 'Lugar de nacimiento', 'Ciudad de residencia'], axis=1, inplace=True)
```

## Imputacion con KNN


```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Load your dataset
# Assuming df is your DataFrame
# df = pd.read_csv('your_dataset.csv')

# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns

# Initialize the KNNImputer with more neighbors
imputer = KNNImputer(n_neighbors=15)

# Fit and transform the numeric data
imputed_numeric_data = imputer.fit_transform(df[numeric_cols])

# Convert the imputed numeric data back to a DataFrame
imputed_numeric_df = pd.DataFrame(imputed_numeric_data, columns=numeric_cols)

# Combine the imputed numeric columns with the non-numeric columns
imputed_df = pd.concat([imputed_numeric_df, df[non_numeric_cols]], axis=1)

# Ensure the original data types are preserved
for column in df.columns:
    imputed_df[column] = imputed_df[column].astype(df[column].dtype)

# Check for any remaining null values
null_counts = imputed_df.isnull().sum()
print(null_counts[null_counts > 0])

# If there are still null values, handle them separately
for col in null_counts[null_counts > 0].index:
    if imputed_df[col].dtype == 'object' or pd.api.types.is_categorical_dtype(imputed_df[col]):
        imputed_df[col].fillna(imputed_df[col].mode()[0], inplace=True)
    else:
        imputed_df[col].fillna(imputed_df[col].mean(), inplace=True)

```
